```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import matplotlib.gridspec as gridspec #Used for organizing figures
import matplotlib.mlab as mlab
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from scipy.stats import vonmises
%matplotlib inline
```

```python
VM_data = np.array(pd.read_csv("VM_analysis.txt"))
```

```python
fig = plt.figure(figsize=(8, 8))
gs = gridspec.GridSpec(1, 1)

ax = plt.subplot(gs[0, 0]) # row 0, col 0
ax.hist(VM_data[:, 0]*180/np.pi, bins = 50, color = 'green', alpha = 0.2)
ax.hist(VM_data[:, 1]*180/np.pi, bins = 50, color = 'blue', alpha = 0.2)
ax.set_xlabel('Rake')
ax.set_ylabel('Count')

plt.show()

```

```python
fig = plt.figure(figsize=(12, 12))
gs = gridspec.GridSpec(2, 2)



ax1 = plt.subplot(gs[1, 0]) # row 0, col 0
ax1.hist2d(VM_data[:, 2]*180/np.pi, VM_data[:, 0]*180/np.pi, bins = 250)
ax1.set_xlabel('Dip')
ax1.set_ylabel('Predicted Rake')



ax2 = plt.subplot(gs[0, 0]) # row 0, col 0
ax2.hist(VM_data[:, 2]*180/np.pi, bins = 250, alpha = 0.2)
ax2.ticklabel_format(style='sci', axis='y', scilimits=(0,0))


ax3 = plt.subplot(gs[1, 1]) # row 0, col 0
ax3.hist(VM_data[:, 0]*180/np.pi, bins = 250, orientation='horizontal', alpha = 0.2, color = "green")
ax3.set_xlabel('Predicted Rakes')
ax3.set_ylabel('Count')
ax3.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

plt.show()


```
