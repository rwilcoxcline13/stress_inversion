```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model, datasets
%matplotlib inline
```

```python
yeti_social_data = np.loadtxt('Lab06_Data.csv', delimiter=',')

age = yeti_social_data[:, 0]
socialization = yeti_social_data[:, 1]
```

```python
def linear_reg_LSQ(data_context, data):
    
    G = np.ones((len(data_context), 2))
    G[:, 0] = data_context
    G_T = G.T
    G_gen = G_T@G
    m = (np.linalg.inv(G_gen)@(G_T@data))
    
    return m
    
    
```

```python
m1, m2 = linear_reg_LSQ(age, socialization)

x_grid = np.arange(10, 18, 0.1)
model = m1*x_grid+m2

d_pred = m1*age+m2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(age, socialization, alpha = 0.5)
ax.plot(x_grid, model, linestyle = '-', color = 'red', linewidth = '1')
```

```python
res = d_pred-socialization
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(np.arange(len(res)), res, alpha = 0.5)
ax.hlines(0, 0, 10, colors='k', linestyles='dashed')
ax.set_xlim([-.075, 10])
```
